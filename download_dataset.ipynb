```python
import kagglehub
import pandas as pd
import os
from pathlib import Path


print("Downloading dataset...")
path = kagglehub.dataset_download("dhoogla/unswnb15")
print("Path to dataset files:", path)

```

```python

output_dir = Path.cwd() / "dataset"
output_dir.mkdir(parents=True, exist_ok=True)


print("\nConverting parquet files to CSV...")


training_file = os.path.join(path, "UNSW_NB15_training-set.parquet")
if os.path.exists(training_file):
    print(f"Reading {training_file}...")
    df_train = pd.read_parquet(training_file)
    output_train = output_dir / "UNSW_NB15_train_40k.csv"
    df_train.to_csv(output_train, index=False)
    print(f"Saved: {output_train}")
else:
    print(f"Warning: {training_file} not found")

# Testing set
testing_file = os.path.join(path, "UNSW_NB15_testing-set.parquet")
if os.path.exists(testing_file):
    print(f"Reading {testing_file}...")
    df_test = pd.read_parquet(testing_file)
    output_test = output_dir / "UNSW_NB15_test_10k.csv"
    df_test.to_csv(output_test, index=False)
    print(f"Saved: {output_test}")
else:
    print(f"Warning: {testing_file} not found")

print("\nConversion complete!")
print(f"Files saved in: {output_dir}")

```
